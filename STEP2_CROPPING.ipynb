```python
import cv2
import numpy as np
import os
from google.colab.patches import cv2_imshow
from PIL import Image
```

TRAIN

```python
# Define the directory containing the images
image_dir = '/content/drive/MyDrive/PPMI/SN_FINAL_ForCrop/Train/Control/'

# Define the directory where cropped images will be saved
output_dir = '/content/drive/MyDrive/PPMI/SN_FINAL_Cropped/Train/Control/'

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Create a list of image file names in the directory
image_files = [f for f in os.listdir(image_dir) if f.endswith('.png')]

# Loop through each image file
for image_file in image_files:
    # Load the image
    image = Image.open(os.path.join(image_dir, image_file))

    # Define the cropping coordinates
    x = 450
    y = 500
    h = 1450
    w = 1150

    # Crop the region of interest
    crop_image = image.crop((x, y, h, w))

    # Generate the output file path for the cropped image
    output_file = os.path.join(output_dir, f'cropped_{image_file}')

    # Save the cropped image using the .save method
    crop_image.save(output_file)

    print(f"Cropped image saved as '{output_file}'")
```
